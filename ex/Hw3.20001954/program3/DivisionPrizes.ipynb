```python
def division_prize(m, n):
    if m == 0 or n == 1:
        return 1
    elif m < n:
        return division_prize(m, m)
    else:
        return division_prize(m, n - 1) + division_prize(m - n, n)
```

```python
print(division_prize(7, 4))
```

```python
import timeit
import matplotlib.pyplot as plt
n_values = [10, 20, 30, 40, 50]
m_values = [20, 40, 60, 80, 100]
times = []
print("Value from each n")
for index in range(len(n_values)):
    t = timeit.timeit(lambda: division_prize(m_values[index], n_values[index]), number = 1)
    times.append(t)

plt.plot(n_values, times)
plt.xlabel("Kích thước n")
plt.ylabel("Thời gian thực hiện (giây)")
plt.show()
```
