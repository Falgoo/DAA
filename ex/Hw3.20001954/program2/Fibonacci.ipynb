# Fibonacci function

```python
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
```

# Plot graph

```python
import timeit
import matplotlib.pyplot as plt
n_values = [5, 10, 15, 20, 25, 30, 35, 40]
times = []

for n in n_values:
    t = timeit.timeit(lambda: fibonacci(n), number = 1)
    times.append(t)

plt.plot(n_values, times)
plt.xlabel("Kích thước n")
plt.ylabel("Thời gian thực hiện (giây)")
plt.show()
```
