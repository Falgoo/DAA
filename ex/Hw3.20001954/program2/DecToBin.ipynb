# Function convert decimal to binary

```python
def decToBin(n):
    if (n > 0):
        decToBin(int(n/2))
        print(n % 2, end="")
```

```python
import timeit
import matplotlib.pyplot as plt
n_values = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000]
times = []
print("Value from each n")
for n in n_values:
    t = timeit.timeit(lambda: decToBin(n), number = 1)
    print()
    times.append(t)

plt.plot(n_values, times)
plt.xlabel("Kích thước n")
plt.ylabel("Thời gian thực hiện (giây)")
plt.show()
```
