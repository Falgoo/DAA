# Function convert decimal to product of primes

```python
def DecimalToProductPrimes(n, divisor=3):
    if (n >= 2):
        if (n % 2 == 0):
            n = n / 2
            print(2, end="*") if n >= 2 else print(2)
        elif (n % divisor == 0):
            n = n / divisor
            print(divisor, end="*") if n >= 2 else print(divisor)
        else:
            divisor += 2
        DecimalToProductPrimes(n, divisor)
```

```python
import timeit
import matplotlib.pyplot as plt
n_values = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000]
times = []
print("Value from each n")
for n in n_values:
    t = timeit.timeit(lambda: DecimalToProductPrimes(n), number = 1)
    times.append(t)

plt.plot(n_values, times)
plt.xlabel("Kích thước n")
plt.ylabel("Thời gian thực hiện (giây)")
plt.show()
```
