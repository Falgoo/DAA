# Import library

```python
import numpy as np
import timeit
import matplotlib.pyplot as plt
```

# Selection sort

```python
def selectionSort(array):
    n = len(array)
    for i in range(n - 1):
        minIndex = i
        for j in range(i + 1, n):
            if array[j] < array[minIndex]:
                minIndex = j
        array[i], array[minIndex] = array[minIndex], array[i]
```

# Calculate the execution time and plot the time dependence on the size

```python
n_values = [100, 200, 500, 1000, 2000, 5000, 10000]
times = []

for n in n_values:
    arr = np.random.randint(1, 1000, size = n)
    t = timeit.timeit(lambda: selectionSort(arr), number = 1)
    times.append(t)

plt.plot(n_values, times)
plt.xlabel("Kích thước n")
plt.ylabel("Thời gian thực hiện (giây)")
plt.show()
```
