# Import library

```python
import numpy as np
import timeit
import matplotlib.pyplot as plt
```

# Majority function

```python
def findMajority(arr, left, right):
    if left == right:
        return arr[left]
  
    middle = (left + right) // 2
  
    left_majority = findMajority(arr, left, middle)
    right_majority = findMajority(arr, middle+1, right)
  
    if left_majority == right_majority:
        return left_majority
  
    left_count = sum(1 for i in range(left, right+1) if arr[i] == left_majority)
    right_count = sum(1 for i in range(left, right+1) if arr[i] == right_majority)
  
    if left_count > (right - left + 1) // 2:
        return left_majority
    elif right_count > (right - left + 1) // 2:
        return right_majority
    else:
        return None
    
def find_majority(array):
    return findMajority(array, 0, len(array) - 1)
```

# Test find_majority function

```python
array = np.random.randint(1, 3, size = 10)

print("Array is:",array)
print("Majority in array is:", find_majority(array))
```

# Calculate the execution time and plot the time dependence on the size

```python
n_values = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000]
times = []

for n in n_values:
    array = np.random.randint(1, 100, size = n)
    array.sort()
    x = np.random.randint(1, 100)
    t = timeit.timeit(lambda: find_majority(array), number = 1)
    times.append(t)

plt.plot(n_values, times)
plt.xlabel("Kích thước n")
plt.ylabel("Thời gian thực hiện (giây)")
plt.show()
```
