# Import library

```python
import numpy as np
import timeit
import matplotlib.pyplot as plt
```

# Binary search

```python
def binarySearch(array, left, right, x):
    if left >= right:
        return -1
    
    middle = left + (right - left) // 2
    
    if array[middle] == x:
        return middle
    elif array[middle] < x:
        return binarySearch(array, middle + 1, right, x)
    else:
        return binarySearch(array, left, middle, x)
    
def binary_search(array, x):
    return binarySearch(array, 0, len(array) - 1, x)
```

# test function binary search

```python
array = np.random.randint(1, 20, size = 10)
x = np.random.randint(1, 20)
array.sort()

print("Array is:",array)
print("X is:",x)
print("Position of",x,"in array:", binary_search(array, x))
```

# Calculate the execution time and plot the time dependence on the size

```python
n_values = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000]
times = []

for n in n_values:
    array = np.random.randint(1, n, size = n)
    array.sort()
    x = np.random.randint(1, 100)
    t = timeit.timeit(lambda: binary_search(array, x), number = 1)
    times.append(t)

plt.plot(n_values, times)
plt.xlabel("Kích thước n")
plt.ylabel("Thời gian thực hiện (giây)")
plt.show()
```
