# Import library

```python
import numpy as np
import timeit
import matplotlib.pyplot as plt
```

# Median sort arrays

```python
def medianSortedArrays(array1, array2):
    n1, n2 = len(array1), len(array2)
    if n1 > n2:
        array1, array2, n1, n2 = array2, array1, n2, n1
    
    if (n1 + n2) % 2 == 1:
        return find_middle(array1, array2, (n1 + n2 + 1) // 2)
    else:
        return (find_middle(array1, array2, (n1 + n2) // 2) + find_middle(array1, array2, (n1 + n2) // 2 + 1)) / 2

def find_middle(array1, array2, middle):
    n1, n2 = len(array1), len(array2)
    if n1 == 0:
        return array2[middle - 1]
    if middle == 1:
        return min(array1[0], array2[0])

    index1, index2 = min(n1, middle // 2), min(n2, middle // 2)
    if array1[index1 - 1] < array2[index2 - 1]:
        return find_middle(array1[index1:], array2, middle - index1)
    else:
        return find_middle(array1, array2[index2:], middle - index2)
```

# Test medianSortedArrays function

```python
arrayA = np.random.randint(1, 10, size = 10)
arrayB = np.random.randint(1, 100, size = 8)
arrayA.sort()
arrayB.sort()

print("Array a:",arrayA)
print("Array b:",arrayB)
print("Median of two array is:", medianSortedArrays(arrayA, arrayB))
```

# Calculate the execution time and plot the time dependence on the size

```python
n_values = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000]
times = []

for n in n_values:
    arrayA = np.random.randint(1, 100, size = n)
    arrayB = np.random.randint(1, 100, size = n-1)
    arrayA.sort()
    arrayB.sort()
    t = timeit.timeit(lambda: medianSortedArrays(arrayA, arrayB), number = 1)
    times.append(t)

plt.plot(n_values, times)
plt.xlabel("Kích thước n")
plt.ylabel("Thời gian thực hiện (giây)")
plt.show()
 
```
