# Import library

```python
import numpy as np
import timeit
import matplotlib.pyplot as plt
```

# Find min max

```python
def findMinMax(array, left, right):
    if left == right:
        return array[left], array[right]
    
    if right - left == 1:
        if array[left] < array[right]:
            return (array[left], array[right])
        else:
            return (array[right], array[left])

    mid = (left + right) // 2
    left_min, left_max = findMinMax(array, left, mid)
    right_min, right_max = findMinMax(array, mid + 1, right)
    
    return min(left_min, right_min), max(left_max, right_max)

def find_min_max(array):
    return findMinMax(array, 0 , len(array) - 1)
```

# Test find_min_max function

```python
array = np.random.randint(1, 10, size = 10)

print("Array is:", array)
print("Min and max of array is:", find_min_max(array))
```

# Calculate the execution time and plot the time dependence on the size

```python
n_values = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000]
times = []

for n in n_values:
    array = np.random.randint(1, 100, size = n)
    x = np.random.randint(1, 100)
    t = timeit.timeit(lambda: find_min_max(array), number = 1)
    times.append(t)

plt.plot(n_values, times)
plt.xlabel("Kích thước n")
plt.ylabel("Thời gian thực hiện (giây)")
plt.show()
```
